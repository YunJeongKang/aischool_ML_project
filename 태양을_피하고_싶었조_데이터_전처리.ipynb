```python
import pandas as pd
import datetime

day_electricity_df = pd.read_csv('1차_프로젝트/day_data/day_electricity_20200101_20220818.csv', encoding="utf-8-sig")
day_electricity_df_time = day_electricity_df["시간"].to_list()
day_electricity_df = day_electricity_df.drop(columns=["시간"])
for i in range(len(day_electricity_df_time)):
    time = datetime.datetime.strptime(day_electricity_df_time[i], "%Y-%m-%d %H:%M")
    time = time.strftime("%Y-%m-%d %H:%M")
    day_electricity_df_time[i] = time
time_df = pd.DataFrame(day_electricity_df_time, columns=["일시"])
day_electricity_df = pd.concat([day_electricity_df, time_df], axis = 1)

day_weather_df2 = pd.read_csv('1차_프로젝트/day_data/day_weather_20200101_20200818.csv', encoding="cp949")
day_weather_df3 = pd.read_csv('1차_프로젝트/day_data/day_weather_20200819_20210818.csv', encoding="cp949")
day_weather_df4 = pd.read_csv('1차_프로젝트/day_data/day_weather_20210819_20220818.csv', encoding="cp949")
day_weather_df = pd.concat([day_weather_df2, day_weather_df3, day_weather_df4], axis = 0)

day_PM_df = pd.read_csv('1차_프로젝트/day_data/day_PM_202001010000_202208182300.csv', encoding="utf-8-sig")

day_train_df = pd.merge(day_electricity_df, day_weather_df, on="일시", how='outer')
day_train_df = pd.merge(day_train_df, day_PM_df, on="일시", how='outer')
for i in range(1, len(day_train_df)):
    if "-" not in day_train_df.loc[i, '미세먼지']:
        if int(day_train_df.loc[i, '미세먼지']) > 80:
            day_train_df.loc[i, '미세먼지'] = 0
        else:
            day_train_df.loc[i, '미세먼지'] = 1
    else:
        day_train_df = day_train_df.drop(i, axis = 0)
        
day_train_df["강수량(mm)"] = day_train_df["강수량(mm)"].fillna(0)
day_train_df = day_train_df.dropna()

day_train_df.to_csv('1차_프로젝트/day_data/day_train.csv',encoding='cp949')
day_train_df
```

```python
# 일출 일몰
import csv
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions
from selenium.webdriver.support.ui import WebDriverWait
import time
import requests
from datetime import date, timedelta, datetime
from bs4 import BeautifulSoup
import re
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
city_list = [
        "강원도", 
        "경기도", 
        "경상남도", 
        "경상북도", 
        "광주광역시", 
        "대구광역시", 
        "대전광역시", 
        "부산광역시", 
        "서울특별시", 
        "세종특별시", 
        "울산광역시", 
        "인천광역시", 
        "전라남도", 
        "전라북도", 
        "제주도", 
        "충청남도", 
        "충청북도"
    ]
area_num = [42, 41, 48, 47, 29, 27, 30, 26, 11, 50, 31, 28, 46, 45, 49, 44, 43]
for n in range(len(area_num)):
    html = requests.get(f"https://www.accuweather.com/ko/browse-locations/asi/kr/{area_num[n]}", headers={"User-Agent" : "Mozilla/5.0"}).text
    soup = BeautifulSoup(html, "html.parser")
    세부_list = soup.find_all("a", {"class":"search-result"})
    area_df = pd.DataFrame(columns={"지역명", "지역번호"})
    num = re.compile('[^0-9+]')
    for i in range(1, len(세부_list)):
        area_list = []
        area_list.append(세부_list[i].text)
        area_list.append(num.sub('', str(세부_list[i]))[:-1])
        area_df = area_df.append(pd.Series(area_list, index=["지역명", "지역번호"]), ignore_index=True)
    area_df.to_csv(f'1차_프로젝트/day_data/{city_list[n]}.csv',encoding='cp949')
```
